```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("/Volumes/RAJ SANKAR/IPL Ball by Ball/ALL_2021_IPL_MATCHES_BALL_BY_BALL.csv")
df.head()
```

```python
df['over'] = np.ceil(df['ball'])
grouped_vls =  df.groupby(['over'],as_index=False).total_runs.sum()
```

```python
grouped_vls =  df.groupby(['over'],as_index=False).total_runs.count()
grouped_vls1 = df.groupby(['over'],as_index=False).total_runs.sum()
```

```python
grouped_vls['total_runs'] = grouped_vls['total_runs'] / 6.0
grouped_vls['avergae'] = grouped_vls1['total_runs'] / grouped_vls['total_runs']
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20, 12), dpi=80)
import matplotlib.pyplot as plt
import seaborn as sns
plt.bar(grouped_vls['over'], grouped_vls['avergae'],color='g')
```
